# Exploratory Data Analysis

Before analysing the data. Any data analysis project should start with a business foundation; a goal that needs to be met to provide business value. In this case, we are starting from a data set and are required to step back and determine if there is business value that may be extracted.

From the documentation, the data set provides data on a series of physicochemical properties of Portuguese Vinho Verde red wine and an associated quality score. With the associated quality determined by the median subjective score of at least 3 experts. The features and output variables are: 

Input variables (based on physicochemical tests):
   1. fixed acidity (tartaric acid - g / dm^3)
   2. volatile acidity (acetic acid - g / dm^3)
   3. citric acid (g / dm^3)
   4. residual sugar (g / dm^3)
   5. chlorides (sodium chloride - g / dm^3
   6. free sulfur dioxide (mg / dm^3)
   7. total sulfur dioxide (mg / dm^3)
   8. density (g / cm^3)
   9. pH
   10. sulphates (potassium sulphate - g / dm3)
   11. alcohol (% by volume)
   
   
   Output variable (based on sensory data): 
   12. quality (score between 0 and 10)


To improve the quality of a wine, a viticulturist may want to know the physical properties of a wine that is associated with higher quality wine. From this insight they may be able to make changes to processes, tradeoffs in costs, as well as customer taste preferences This hypothetically, would allow a viticulturist to help optimise their product design and selection.

For this exercise, this notebook will explore the physicochemical properties that are associated with a high quality of Portuguese Vinho Verde wine.
After the exploration and data analysis phase. A machine learning model will be constructed to help predict the quality of the wine. 

For ease of readability, this is completed over a series of notebooks each with its focal points and an application folder for a web application

### Linear Correlation

To better gauge the relationships between the features, we can use a correlation matrix to understand the linear relationships. 

We can see that a higher quality of the wine is linearly positively correlated to high alcohol content, sulphates, and citric acid. But linearly negatively correlated with volatile acidity, total sulfur dioxide, and density. Further, we can see density is strongly correlated with fixed acidity. 

Whilst being correlated, the correlation may be spurious or unimportant. Before spending time understanding what each of these represents. It is usually worthwhile building a statistical model and determining feature importance. Then concentrating on the most important features.

```python
plt.figure(figsize = [14,12])
sns.heatmap(df_raw.corr(),vmin = -1, vmax = 1, cmap = 'coolwarm_r',annot = True)
# plt.title("Correlation Matrix\n", size = 30)
plt.savefig("linear_correlation.jpeg")
```

### Non-Linear Correlation (Rank)
To better gauge how the variables are related to each other in a non-linear sense, a spearman's rank correlation can be used. To better visualise the spearmans rank, a dendrogram can be used. The  dendrogram can be interpreted by the closeness of the links between each variable to the right. The closer the links are to the to the right, the more correlated the variables.

From the previous analysis, the results are generally unsurprising. 
It is found that 
- sulfur and free sulfur are highly correlated. 
- citric acid and fixed acid are highly correlated, though interestingly neither are highly correlated to volatile acidity
- Sulphates to be somewhat correlated to alcohol

Though surprisingly alcohol and residual sugar are uncorrelated. Without a deep understanding of viticulture. One would typically expect alcohol to a function of the fermentation of sugars. This may suggest residual sugar and alcohol are relatively correlated (either negatively as sugar if heavily fermented or positively if more saturation of sugar leads to more alcohol). However, this is not the case.


```python
corr = np.round(scipy.stats.spearmanr(df_raw).correlation, 4)
corr_condensed = hc.distance.squareform(1-corr)
z = hc.linkage(corr_condensed, method='average')
fig = plt.figure(figsize=(16,10))
dendrogram = hc.dendrogram(z, labels=df_raw.columns, orientation='left', leaf_font_size=16) #Change font size to 11.5 for images to stop label cropping
plt.title("Spearmans Rank Correlation\n",size = 30);
# plt.savefig("Spearmans Rank Correlation.jpeg") 
```

### Quality Analysis

We can see the quality is somewhat normally distributed. Though there is a slight skew to the right. For linear models, this may require some adjustments. 

Further we can see that the counts of 3, 4 and 8 are considerably lower than that of the other groups. This is an issue as it can lead to misclassification in these classes. 
Given there isn't an option to get more data in these classes. We may need to rely on resampling techniques in the test set such as boosting (sampling with replacement with higher weights on misclassified examples), or some other form of oversampling from the underrepresented classes. 

Another issue with the data is that the dependant variable is not completely represented in its own space.   

Lets define the population of all possible values to be the set S and the set of samples provided to be E. Such that E $\subset$ S. According to the data set description, the complete space is bounded below by 0 and above by 10. If we take the supremum and infimum over the set E, we get the values 3 and 8 respectively (min and max of the subset in this case). However given that the supremum and infimum are not unique in their upper and lower bounds sets (where ub = {1,2,3}, lb = {8,9,10}). Then we must extrapolate from the data into this space. In order to be able to predict cases where events lie in this space without having any data associated with the event, it is required that a model that is capable of extrapolating into these spaces be used. Otherwise the dependant variable may be further discritised into a smaller space i.e S = {"poor","medium","good"}, where each label contains values already used. This makes is difficult for non parametric models such as CART, KNN etc as they can't extrapolate into unseen space. As an alternative, Parametric models such as a Linear/Logistic model or Neural Network could potentially be trained and constained to an output space between 0 and 10 with a regression model (i.e with a sigmoid function), or be constrained to discrete classes (with a softmax function). 

```python
plt.figure(figsize = [15,10])
plt.title("Quality Analysis\n",size=30)
plt.ylabel("Count",size=20)
plt.xlabel("Quality Score",size=20)
df_raw["quality"].hist(grid = False);
# plt.savefig("Quality Histogram.jpeg")
```

#### Free Sulphur Dioxide
The Free Sulphur Dioxide appears to be quadraticaly related to the quality of the wine.

This quadratic relationship may make sense if there is a tradeoff between microbial growth and oxidation say in fermentation and the freshness of the wine.
